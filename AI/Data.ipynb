```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Snipers.csv')

# データの相関関係を計算
correlation_matrix = df.corr()

# ヒートマップを作成
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# CSVファイルを読み込む
df = pd.read_csv('Snipers.csv')

df2 = pd.read_csv('Data.csv')

# bedwars_levelが30以下のデータを抽出
filtered_data = df2[(df2['bedwars_level'] <= 120) & (df2['Cheat'] == 1)]

# ランダムに30個選択
df2 = filtered_data.sample(n=50)

# ラベルを変数にしておく
x_label = 'bedwars_level'
y_label = 'beds_lost_bedwars'

title = 'BEDWARS DATA'
# グラフ化
# plt.figure(figsize=(10, 6))
# for cheat in [0, 1]:
#     mean_values = mean_fall_deaths_bedwars.loc[:, cheat]
#     plt.plot(mean_values.index, mean_values.values, marker='o', label=f'Cheat = {not bool(cheat)}')
plt.figure(figsize=(15, 8))
# plt.scatter(df[x_label], df[y_label], s=5, color='blue')
# plt.scatter(df2[x_label], df2[y_label], s=5, color='red')

# x軸ごとにデータをグループ化して平均を計算
mean_data = df.groupby(x_label)[y_label].mean()
plt.plot(mean_data.index, mean_data, color='blue')

# x軸ごとにデータをグループ化して平均を計算
mean_data2 = df2.groupby(x_label)[y_label].mean()
# 平均を棒グラフで表示
plt.plot(mean_data2.index, mean_data2, color='red')
# 平均の線を引く
# plt.axvline(mean_a, color='b', linestyle='--', label='Mean of ColumnA')

plt.xlabel(f"{x_label}")
plt.ylabel(y_label)
plt.title(title)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# CSVファイルを読み込む
df = pd.read_csv('Data.csv')

# 特徴量を正規化する
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)

# 正規化したデータフレームを作成
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)

# ヒートマップを描画
plt.figure(figsize=(12, 8))
sns.heatmap(scaled_df.corr(), fmt=".2f")
plt.title('Correlation Heatmap (Standardized Data)')
plt.show()
```
